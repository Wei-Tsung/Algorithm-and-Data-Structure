```python

# showing line number to debug , important tips

class MyLinkedList:
    
    def __init__(self,val=None):
        
         
        self.val = val
        self.next = None
        
#get沒有問題     
    def get(self,index):
        
        if self.val == None: #head
            return -1
        
        if index == 0:
            return self.val
        
        node = self.next 
        i = 1
        
        while node != None:
            if i == index:
                return node.val
            
            
            i = i + 1
            node = node.next
                 
        return -1
        
    def addAtHead(self, val): #測試過沒有問題
            
        if self.val == None:  
            self.val = val
            return 
        
        
        thir_node = self.next  # 變動連結的順序和值  而不是變動記憶體位址
        
        temp = self.val
        self.val = val
        
        sec_node = MyLinkedList(temp)
        
         
        
        self.next = sec_node
        sec_node.next = thir_node
        
        
        return  
            
    def addAtTail(self, val): #完成
        
        
        if self.val == None :
            
            self.val = val
            
            return
               
        node = self # 注意這裡的寫法1. 位置不同 意義也不同
        
        while node.next != None: #注意這裡的寫法2. 位置不同 意義也不同
            
            node = node.next
                   
    
        
        tail = MyLinkedList(val)
        node.next = tail
        
        return
    
    
    def addAtIndex(self, index , val) : #這部分還沒有檢查過
        
        if index <= 0:
            self.addAtHead(val)
            return 
        
        prev = self
        node = self.next
        i = 0
        
        while node != None : 
            
            if i == (index-1): # 在插入的位置的前一個節點就必須要連接
                
                new = MyLinkedList(val)    
                prev.next = new
                new.next = node 
                
                return 
            
            prev = prev.next
            node = node.next
            i = i + 1
        
        if (i+1) < index :
            return
        
        elif (i+1) == 0 :
            self.addAtTail(val)
            return 
        
    def deleteAtIndex(self, index): # 
       # 尚未了解清楚 
        if (index < 0):
            return
        
        if (index == 0):
            if self.next == None :
                self = None
                return # 回傳了之後以下就不執行了
            
            self.val = self.next.val # 變更首項的值 而不是轉變所有的位址的連接方式
            self.next = self.next.next # 指標指向到原本的第3個數  非常有趣的地方
            return
        #以上刪除第一個linkedlist節點數值的情況  , 例外處理 
        #self 首個節點
        
       #重新來過
    
        prev = self
        node = self.next
        i = 0
        while (i < index) :
                
            if prev == None:
                return
                
            if i == (index-1):
                prev.next = node.next
                return
            prev = prev.next
            node = node.next
            i = i + 1
        
        
    
     
```

```python
a = MyLinkedList(1)
b = MyLinkedList(2)
c = MyLinkedList(3)
d = MyLinkedList(4)


a.next = b
b.next = c
c.next = d

```
