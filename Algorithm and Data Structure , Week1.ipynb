```python
#size如何使用?

class MyLinkedList:
    
    def __init__(self,val=None):
        
        self.head = self    
        self.val = val
        self.next = None
         
        
         
    def get(self, index :int): # get 寫完整
        
        
        
        if self.val is None:
            return -1
        
        node = self
        i = 0
        while node != None :   # i會變動逼近  index輸入時已經固定
            
            if i == index:
                return node.val # 這裡輸出要了解清楚
            
            node = node.next  # index的重要性
            i = i + 1
            
                     
                
            
             
        return -1 # 其他一切的例外處理
        

  
    def addAtHead(self, val): # addAtHead目前沒有問題
        """
        Add a node of value val before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
        :type val: int
        :rtype: None
        """
        if self.head == None :
            self.val = val #如果第一個是空值  重置成head
            self.head = val 
            return #回傳
        
        
        new_head =  MyLinkedList(val)  #先建立新的node 使用Mylinklist 建立reference物件
        new_head.next = self
        self.head = new_head
        
        return self.head           
    
    
    def addAtTail(self, val):
        """
        Append a node of value val to the last element of the linked list.
        :type val: int
        :rtype: None
        """
        if self.val  ==  None:
            self.val = val
             
            return
           
         
        node = self
        #不必在乎index次數  直接到最後一個node    
        while node.next != None :  # 直到next 為None才結束
            node = node.next
            
        new_node = MyLinkedList(val)    
        node.next = new_node    
         
        
        return 
        
    #未完成    
    def addAtIndex(self, index : int , val):
        """
        Add a node of value val before the index-th node in the linked list. If index equals to the length of linked list, the node will be appended to the end of linked list. If index is greater than the length, the node will not be inserted.
        :type index: int
        :type val: int
        :rtype: None
        """
        #頭部和尾巴只需要進行連結的動作
        #其他中間區塊則必須要進行頭尾斷開重新連結的動作
        
         
        #例外先處理
        
        
        if index <= 0 : # index=0 也是同樣比照<0辦理 
             return self.addAtHead(val) # return new_head 動作極端的關鍵
        
        i = 0 # (目前的index)
        
        prev = self
        node = self.next
        while (node != None) :  # 一直走到linked_list的最後一個
            
            #  正規處理     #Add a node of value val before the index-th node in the linked list
            if i == (index-1) :
                new_node = MyLinkedList(val)
                prev.next = new_node
                new_node.next = node
                return  
                
            prev = prev.next
            node = node.next
            i = i + 1
          
        # i + 1 -> 總長度
        if (i+1) == index : 
            return self.addAtTail(val)      
            
        if (i+1)  < index :  #will not be inserted
            return         
            
    def deleteAtIndex(self, index : int):
        
        """
        Delete the index-th node in the linked list, if the index is valid.
        :type index: int
        :rtype: None
        """
        
        prev = self 
        node = self.next # 想要移除的節點
        
        if (self.val == None) or (index < 0):
            return -1
        
        if index == 0 :  
            self = self.next
        
        i = 0
        
        while node != None : #想要操作的對象是什麼?
            
            
            if node.next == None:
                prev.next = None
                return 
            
            if i == index-1 : #index-1 :prev  才是想要操作的對象
                prev.next = node.next
                
            
            prev = prev.next
            node = node.next
            i = i + 1
            
        
        
        return -1 # 其餘的錯誤處理
        
        
       
            
        
             
        
        
          
#小心retrun None 的問題            
#創造一個self.size值會讓code變得更簡潔有效率
            
            
              
            
        
```
